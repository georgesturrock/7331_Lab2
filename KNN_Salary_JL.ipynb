```python
import numpy as np
import pandas as pd
df = pd.read_pickle("./analysisdata/player.pkl")
print('Dataframe Size:',df.shape)
print('Dataframe Null values:',df.isnull().values.ravel().sum())
print ('Number of unique classes:\n',df.nunique())
```

```python
# Printing Head and Tail of Dataset
df.iloc[np.r_[0:6, -6:0]]
```

```python
#Drop Variables with no predictive ability
df.drop(['MinSalaries', 'salary', 'adjSalary', 'adjMinSalaries','playerID'], inplace=True, axis=1)
```

```python
#Creating Dummy variables for categorical variable 'Primary' positions
df2 = pd.get_dummies(df.Primary,prefix='Prim', prefix_sep='_')
df = df.join(df2, how='inner')
df.drop(['Primary'], axis=1, inplace=True)
```

```python
#Set minimum games at 30
df = df[df['G_x'] > 30]
df = df.reset_index(drop=True)
df.iloc[np.r_[0:6, -6:0]]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn import metrics as mt
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')

# create variables we are more familiar with
y = df.adjSalary2
X = df.drop(['adjSalary2'], axis=1)
```

```python
# Scaling training variables
scl = StandardScaler()
X = scl.fit_transform(X)

from sklearn.model_selection import train_test_split
#Create Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Create K-Fold quantity
cv = 10

#KNN regressor 10-fold cross-validation 
from sklearn.neighbors import KNeighborsRegressor

RegEstimator = KNeighborsRegressor()
parameters = { 'n_neighbors':[3,5,15,30], 'weights': ['uniform','distance'],'leaf_size': [5,10,30], 'metric': ['euclidean','minkowski']}

#Create a grid search object using the  
from sklearn.model_selection import GridSearchCV
KNNGridSearch = GridSearchCV(estimator=RegEstimator, n_jobs=6, verbose=1, param_grid=parameters, cv=cv, scoring='r2')

#Perform hyperparameter search to find the best combination of parameters for our data
KNNGridSearch.fit(X_train, y_train)
```

```python
print("Best Estimator:", KNNGridSearch.best_estimator_)
print("Best Parameters:", KNNGridSearch.best_params_)
print("Best Score:", KNNGridSearch.best_score_)
```

##### Removing Highly Correlated Variables
In an attempt to improve R2, Investigate removing highly correlated variables and high VIF

```python
X = df.drop(['adjSalary2'], axis=1)
y = df.adjSalary2

#Drop highly correlated, insignificant and high VIF columns.
X = X.drop(['H', 'G_y','G_x','InnOuts','HR','GS','AB','SB_y','stint','A','R','Prim_1B','OBP','FPct','Bavg','Slug','debutYear'], axis=1)

#Create correlation matrix
CorrMat = X.corr()

# Highest Correlation Pairs
corrPairs = CorrMat.unstack().sort_values(kind="quicksort")
#- REMOVE DUPLICATES
corrPairs = corrPairs[::2]
corrPairs = corrPairs[corrPairs.index.get_level_values(0) != corrPairs.index.get_level_values(1)]
with pd.option_context('display.max_rows',16):
    print(corrPairs)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

df2_vif = pd.DataFrame()
df2_vif["VIF Factor"] = [vif(X.values, i) for i in range(X.shape[1])]
#df2_vif["features"] = df_teamX_scaled.columns
df2_vif["features"] = X.columns
df2_vif
```

```python
# Scaling training variables
scl = StandardScaler()
X = scl.fit_transform(X)

#Create Cross Validation Object with 10 folds with 80/20 train - test split
cv = ShuffleSplit(n_splits = 10, test_size=0.20, random_state=0)

# now iterate through and get predictions, saved to the correct row in yhat

for train, test in cv.split(X,y):
    # get a handle to the classifier object, which defines the type
    reg_knn = KNeighborsRegressor(n_neighbors=15, weights='distance', metric='euclidean',algorithm='auto',leaf_size=5)
    reg_knn.fit(X[train],y[train])
    coeffDeterm = reg_knn.score(X[test],y[test])
    print(coeffDeterm)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(reg_knn, X, y, scoring='r2')
```
