### Important Attributes for Position Classification

The modeling techniques used for the position classification task were KNN, LDA, and random forest classification models. KNN uses neighboring points classifications; therefore, feature importance is not a valid measure. Linear Discriminant Analysis is dimensionality reduction technique. As a result, interpreting the most significant attributes in the model is very complicated, since modeling variables are combinations of the original variables. So in order to determine the most important features in classifying player's positions, we will examine the feature importance plot from the random forest models. 

```python
from IPython.display import Image
Image(filename = "./Feature Importance/RF1.png")
```

The above plot shows the feature importance plot from the most accurate random forest model. The darker the box, the more important the attribute is classifying position. From the plot, the most important attributes are DP, or number of double plays, and A, or number of assists. The attributes P, or putouts, is slightly less significant than the number of double plays and assists, but still very important in determining position. SB_y, or the number of stolen bases, is the fourth most important attribute in determining position. CS_y, or the number of times caught stealing, is the fifth most important followed by FPct, fielding percentage, and G_y, the number of games played. The rest of the attributes added minimal or no value to classifying player's positions.

The most majority of the most important attributes found above are defensive statistics. This makes intuitive sense as the player's positions are often determined by how well they can field the ball, throw the ball, etc. For example, if a player has a lot of assists or double plays, then they are more likely to be in the infield.
